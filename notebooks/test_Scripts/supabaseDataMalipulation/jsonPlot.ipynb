```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load JSON
with open("binning_data.json", "r") as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert ms timestamps to datetime
df["start_time"] = pd.to_datetime(df["start_time"], unit="ms")
df["end_time"] = pd.to_datetime(df["end_time"], unit="ms")

# Use start_time as the time axis
df = df.sort_values("start_time")
print(df)
```

```python
# Plot
plt.figure(figsize=(12,6))
plt.plot(df["start_time"], df["score"], label="Score", marker="o")

# Shaded range
plt.fill_between(df["start_time"], df["score_min"], df["score_max"], 
                 alpha=0.3, label="Min-Max Range")

plt.xlabel("Time")
plt.ylabel("Score")
plt.title("Score Over Time with Min-Max Ranges")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
